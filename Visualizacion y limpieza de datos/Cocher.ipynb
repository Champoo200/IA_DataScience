En esta practica haremos multiples visualizaciones de coches por gasto de gasolina dividiendolo en puntos de interes como puede ser marca,etc  

```python
import pandas as pd
import numpy as np

#Abrimos data set para su lectura
car_gas = pd.read_csv("Consumo Gasolina Autos Ene 2018.csv")
car_gas.head()
```

```python
#Realizamos una primera visualizacion de los datos
ford = 'FORD'

Ford_insc = car_gas[car_gas['Marca'] == ford ] 
```

```python
#Importamos altair para visualizar rapidamente los datos

import altair as alts


#Creamos primera grafica


alt.Chart(Ford_insc).mark_bar().encode(
    x='Versión',
    y="Cilindros",
    # The highlight will be set on the result of a conditional statement
    color=alt.condition(
        alt.datum.year == 1810,  # If the year is 1810 this test returns True,
        alt.value('orange'),     # which sets the bar orange.
        alt.value('steelblue')   # And if it's not true it sets the bar steelblue.
    )
).properties(width=600)
```
